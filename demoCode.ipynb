# 1. Linear Restricted Model class

```python
class LinearRestricted(object):
    def __init__(self, mZ, **args):
        # for linear restricted model, we need to provided the predetermined loading matrix
        self.loading = mZ
        self.name = args.get('prefix', '') + 'LinRes'

    def train(self, x, y=None, **args):
        # For linear restricted model, there is no parameters needs to be trained!
        # The "parameters" is the predetermined loading matrix
        # I keep the "train" method here in order to make the api consistent
        return self

    def predict(self, features):
        mF = self.encoding(features)
        return self.decoding(mF)

    def encoding(self, obs):
        nsamples = obs.shape[0]
        feature_size = obs.shape[1]
        
        # here I reshape the data (obs) to 3-D in order to use np.matmul and np.linalg.solve 
        # to solve batch matrix linear systems (aka, slightly faster and avoid loop)
        mY = obs.reshape([nsamples, feature_size, 1])
        mZ = np.repeat(
            self.loading, nsamples,
            axis=0).reshape([nsamples, feature_size, -1], order='F')

        mZt = np.transpose(mZ, (0, 2, 1))
        ZZ = np.matmul(mZt, mZ)
        ZY = np.matmul(mZt, mY)
        mF = np.linalg.solve(ZZ, ZY)
        mF = np.squeeze(mF, 2)
        return mF

    def decoding(self, latent):
        nsamples = latent.shape[0]
        feature_size = self.loading.shape[0]
        mZ = np.repeat(
            self.loading, nsamples,
            axis=0).reshape([nsamples, feature_size, -1], order='F')

        if latent.ndim == 2:
            latent = np.expand_dims(latent, 2)
        fitted = np.matmul(mZ, latent)
        fitted = np.squeeze(fitted, axis=2)
        return fitted

    @property
    def num_params(self):
        return 0
```

#### helper function to extract the loading

```python
def extract_loading(df):
    df['const'] = 1
    df['maturity'] = df['maturity'] / 365 - 0.5
    df['delta'] = df['delta'] / 100 - 0.5
    df['delta2'] = df['delta']**2
    df['inter'] = df['delta'] * df['maturity']
    exogs = ['const','delta','maturity','delta2','inter']
    return df[exogs].astype(np.float32).values
```

```python
inputs = Input(batch_shape=imvol.input_dims, name="input_layer")
inter_encoder = Dense(16, activation='tanh', name='intermediate_encoder_layer')(inputs)
hidden = Dense(5, activation='tanh', name='hidden_layer')(inter_encoder)
inter_decoder = Dense(16,activation='tanh', name='intermediate_decoder_layer')(hidden)
outputs = Dense(imvol.feature_size, name='output_layer')(inter_decoder)

# Build encoder (input: data input, output: hidden layer)
encoder = keras.Model(inputs,hidden)

# Build autoencoder (input: data input, output: reconstruct input (fitted output))
autoencoder = keras.Model(inputs, outputs)

# Build decoder (! here is the tricky part! we need to REUSE the decoder layers in autoencoder)
hidden_input = Input(shape=(5,))
decoder_1 = autoencoder.get_layer('intermediate_decoder_layer')(hidden_input)
decoded = autoencoder.get_layer('output_layer')(decoder_1)
decoder = keras.Model(hidden_input, decoded)
```

The keys for the encoder and decoder (instance of keras.Model) are that they **SHARED** the weights trained by autoencoder
AKA: there is *no new weights need to be trained* by encoder or decoder, they are just part of autoencoder!
We can confirm this by print out the tf.trainable_varialbes(). Keep in mind that keras use Tensorflow as backend, all
the weights (for both kernel (W) and bias (b)) are essentially the Tensorflow Tensor instance 

```python
tf.trainable_variables()
```

As you can see, we only have **FOUR layers** of weightts needed to be trained, they are: intermediate_encoder_layer, hidden_layer
intermediate_decoder_layer, and output_layer. Each of them have a kernel (W) and bias (b) varibles. That's it!

## 2.2 Full feature implementation

Create a abstraction class for sub-classing. (Eaiser to aviod stupid copy and paste same code)

```python
class BaseModel(object):
    '''
    Abstraction Class!
    Don't use without subclassing!
    '''
    def __init__(self, *args):
        pass

    def compile(self,
                optimizer,
                loss=None,
                metrics=None,
                loss_weights=None,
                sample_weight_mode=None,
                weighted_metrics=None,
                target_tensors=None):
        self.model.compile(
            optimizer,
            loss=loss,
            metrics=metrics,
            loss_weights=loss_weights,
            sample_weight_mode=sample_weight_mode,
            weighted_metrics=weighted_metrics,
            target_tensors=target_tensors)

    def train(self,
              x=None,
              y=None,
              batch_size=None,
              epochs=1,
              verbose=1,
              callbacks=None,
              validation_split=0.0,
              validation_data=None,
              shuffle=True,
              class_weight=None,
              sample_weight=None,
              initial_epoch=0,
              steps_per_epoch=None,
              validation_steps=None):

        hist = self.model.fit(
            x=x,
            y=y,
            batch_size=batch_size,
            epochs=epochs,
            verbose=verbose,
            callbacks=callbacks,
            validation_split=validation_split,
            validation_data=validation_data,
            shuffle=shuffle,
            class_weight=class_weight,
            sample_weight=sample_weight,
            initial_epoch=initial_epoch,
            steps_per_epoch=steps_per_epoch,
            validation_steps=validation_steps)
        return hist

    def predict(self, features):
        return self.model.predict(features)

    def evaluate(self, features, labels, verbose=0):
        return self.model.evaluate(features, labels,verbose=verbose)

    @property
    def num_params(self):
        return self.model.count_params()
```

```python
class AutoEncoder(BaseModel):
    def __init__(
        self,
        encoder_units,
        decoder_units,
        hidden_units,
        input_dims,
        output_dims,
        act_fn=None,
        **config
    ):
        '''
        encoder_units: list, specify the encoder units in list. E.g, [32,16,10]
        decoder_units: list, specify the decoder units in list. E.g, [10,16,32]
        hidden_units: integer, the number of hidden units in the bottleneck layer. E.g, 5 (consistent with linear restricted regression 5 factors model)
        input_dims: tuple like, specify the input data feature dimension. E.g, (208,)
        output_dims: tuple like, specify the output dimension. E.g, (208,)
        act_fn: activation function for the hidden layer! NOTE: all intermediate layers implicitly use TANH as activation function. (you can change it in the source code)
        '''
        
        
        if not isinstance(encoder_units, list):
            raise TypeError("encoder units must be a list")
        if not isinstance(decoder_units, list):
            raise TypeError("decoder units must be a list")
        if not isinstance(hidden_units, int):
            raise TypeError("hidden units must be a integer")

        self.encoder_units = encoder_units
        self.decoder_units = decoder_units
        self.hidden_units = hidden_units
        self.input_dims = input_dims
        self.output_dims = output_dims
        
        self.act_fn = act_fn
        self.scope = config.get("scope", "AutoEncoder")
        self.config = config

        self.kernel_regularizer = config.get("kernel_regularizer", None)
        self.bias_regularizer = config.get("bias_regularizer", None)
        self.activity_regularizer = config.get("activity_regularizer", None)

        self.build(input_dims, output_dims)

    def build(self, input_dims, output_dims, **args):
        inputs = Input(batch_shape=input_dims, name=self.scope + "DataInput")
        hidden, self.encoding_layers = self._encoding(inputs)
        decoded, self.decoding_layers = self._decoding(hidden, output_dims)

        self.encoder = keras.Model(inputs, hidden)
        self.model = keras.Model(inputs, decoded)

        encode_input = Input(
            shape=(self.hidden_units,), name=self.scope + "HiddenInput"
        )
        decoded = self._get_decoder(encode_input)
        self.decoder = keras.Model(encode_input, decoded)

    def _get_decoder(self, hidden):
        for layer in self.decoding_layers:
            hidden = layer(hidden)
        return hidden

    def _encoding(self, features):
        encoding_layers = []
        for i, unit in enumerate(self.encoder_units):
            dense = Dense(
                unit,
                activation="tanh", # You can change if you want
                name=self.scope + "Encoder" + str(i),
                kernel_regularizer=self.kernel_regularizer,
                bias_regularizer=self.bias_regularizer,
                activity_regularizer=self.activity_regularizer,
            )
            features = dense(features)
            encoding_layers.append(dense)

        hidden_layer = Dense(
            self.hidden_units,
            activation=self.act_fn,
            name=self.scope + "Hidden",
            kernel_regularizer=self.kernel_regularizer,
            bias_regularizer=self.bias_regularizer,
            activity_regularizer=self.activity_regularizer,
        )

        features = hidden_layer(features)
        encoding_layers.append(hidden_layer)
        return features, encoding_layers

    def _decoding(self, hidden, output_shape):
        decoding_layers = []

        for i, unit in enumerate(self.decoder_units):
            dense = Dense(
                unit,
                activation="tanh", # You can change if you want
                name=self.scope + "Decoder" + str(i),
                kernel_regularizer=self.kernel_regularizer,
                bias_regularizer=self.bias_regularizer,
                activity_regularizer=self.activity_regularizer,
            )
            hidden = dense(hidden)
            decoding_layers.append(dense)

        dense = Dense(
            output_shape[-1],
            name=self.scope + "Decoded",
            kernel_regularizer=self.kernel_regularizer,
            bias_regularizer=self.bias_regularizer,
            activity_regularizer=self.activity_regularizer,
        )

        output = dense(hidden)
        decoding_layers.append(dense)
        return output, decoding_layers

    def encoding(self, data):
        return self.encoder.predict(data)

    def decoding(self, hidden):
        return self.decoder.predict(hidden)
```

# 3. Simple VAR(1)

```python
from statsmodels.tsa.api import VAR
class VAR1(object):
    def __init__(self, **args):
        pass
        
    def train(self, mX, mY=None, **args):
        self.var = VAR(mX).fit(1)

    def predict(self, mX):
        fct = np.zeros_like(mX)
        for i in range(mX.shape[0]):
            fct[i] = self.var.forecast(mX[[i]], 1).ravel()
        return fct
```

```python
# 0. Get data instance
imvol = im.IMVOL('AAPL') 

# 1. Initialize all dimension reduction models (in this script I only implement linear restricted model and autoencoder)
# I build two autoencoder (shadow vs deep) for illustration
mZ = extract_loading(imvol.data.iloc[:imvol.feature_size]) 
linres = LinearRestricted(mZ) 
ae_shadow = AutoEncoder([16,10],[10,16],5,imvol.input_dims, imvol.output_dims) 
ae_deep = AutoEncoder([63,32,16,10],[10,16,32,64],5,imvol.input_dims, imvol.output_dims)
ae_shadow.compile(optimizer='adam',loss='mae')
ae_deep.compile(optimizer='adam',loss='mae')

# 2. Initialize all dynamic models (here we only have VAR)
var1 = VAR1()

# 3. Put all models into dictionary (so that we can easily enumerate)
reductions = {"LinRes":linres, "AE_shadow":ae_shadow, "AE_deep":ae_deep}
transitions = {"VAR1":var1}
```

```python
# 4. train, evaluate, and everyting you want to do
# here, I just compute the FORECASTING error of IVS
for name_reduce, model_reduce in reductions.items():
    model_reduce.train(imvol.train.x, imvol.train.x, validation_data=(imvol.valid.x,imvol.valid.x), batch_size=128,epochs=1000,verbose=0)
    hidden_trainX = model_reduce.encoding(imvol.train.x)
    hidden_trainY = model_reduce.encoding(imvol.train.y)
    hidden_validX = model_reduce.encoding(imvol.valid.x)
    hidden_validY = model_reduce.encoding(imvol.valid.y)
    
    for name_trans, model_trans in transitions.items():
        model_trans.train(hidden_trainX, hidden_trainY,validation_data=(hidden_validX,hidden_validY), batch_size=128,epochs=1000,verbose=0)
        
        # when code runs here, we have our model_reduce trained and model_trans trained
        hidden_test = model_reduce.encoding(imvol.test.x)
        hidden_test_forecast = model_trans.predict(hidden_test)
        output_forecast = model_reduce.decoding(hidden_test_forecast)
        
        print(f"{name_reduce}-{name_trans}: {np.sqrt(np.mean((output_forecast-imvol.test.y)**2))}")
        
        
```

## 5.1 Method 1 Eigen Decomposition

```python
n = 5 #number of components
x_mean = mX.mean(axis=0)
x = mX - x_mean
mCov = np.cov(x,rowvar=0)
eigval, eigvec = np.linalg.eigh(mCov)
eigval, eigvec = eigval[::-1], eigvec[:,::-1] # numpy eigenvalues are in ascending order
mPC, mEv = x@eigvec[:,:n], eigvec[:,:n]
```

```python
plt.plot(mPC)
```

**transform back**

```python
mFit = mPC@mEv.T + x_mean
plt.plot(np.sqrt(np.mean((mX-mFit)**2,axis=0)))
```
